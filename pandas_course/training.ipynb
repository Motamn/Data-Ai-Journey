```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv")
```

```python
df.loc[1:3, "Units Sold"] = 10 # for multiple items 
df.iat[0,0] = "Potato" # for single item
```

```python
df.head(5)
```

```python
df["Units Sold"]
```

```python
df.sort_values("Units Sold")
# df.sort_values("Units Sold", ascending=False) # For Decressing order
df.sort_values(["Units Sold", "Coffee Type"], ascending=[0,1])
```

```python
for index, row in df.iterrows():
  print(index)
  print(row)
  print("\n\n\n")
# this method is slow and inefficiant  
```

```python
for row in df.itertuples():
    print(row.Index, row[1], row[2], row[3])
# Faster that iterrows() but can't use row["Units S"] because it's tuple so you can use the index instead because it's an integer
```
